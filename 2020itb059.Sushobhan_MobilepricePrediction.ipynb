```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Mobile phone price.csv')
df
```

```python
X= df.iloc[:,0:7]

X['Brand'] = pd.factorize(X['Brand'])[0]
X['Model'] = pd.factorize(X['Model'])[0]
```

```python
print(X.columns)
```

```python
X['Storage '] = X['Storage '].str.replace('GB', '')
X['Storage '] = X['Storage '].astype('int64')
X['RAM '] = X['RAM '].str.replace('GB', '')
X['RAM '] = X['RAM '].astype('int64')
```

```python
X.info()
```

```python
for i in range(len(X['Screen Size (inches)'])):
    a=""
    s=X['Screen Size (inches)'][i]
    for j in range(len(s)):
        if(s[j]!=' '):
            a=a+s[j]
        else:
            break
    X['Screen Size (inches)'][i]=a
X['Screen Size (inches)']
```

```python
X['Screen Size (inches)']=X['Screen Size (inches)'].astype(float)
print(X['Screen Size (inches)'].dtype)
```

```python
for i in range(len(X['Camera (MP)'])):
    a=""
    s=X['Camera (MP)'][i]
    
    for j in range(len(s)):
        if(s[j]!='+' and s[j]!='M'):
            a=a+s[j]
        else:
            break
    X['Camera (MP)'][i]=a
X['Camera (MP)']
```

```python
X
```

```python
X['Camera (MP)']=X['Camera (MP)'].astype(float)
```

```python
X.info()
```

```python
Y=df.iloc[:,-1]
Y
```

```python
for i in range(len(Y)):
    a=""
    s=Y[i]
    for j in range(len(s)):
        if(s[j]!='$' and s[j]!=','):
            a=a+s[j]
        elif(s[j]==' '):
            break
            
    Y[i]=a        
print(Y)
```

```python
Y= Y.astype('int64')
Y.dtype
```

```python
X
```

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(X, Y, test_size= 0.25, random_state=0)  
  
# feature Scaling  
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()  
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor 
  
regressor = DecisionTreeRegressor(random_state = 0) 
regressor.fit(x_train,y_train)
```

```python
y_pred_list = []

# Iterate over each row in x_test and predict the corresponding output value
for row in x_test:
    # Reshape the row to have a shape of (1, n_features) to match the expected input shape of the predict method
    row = row.reshape(1, -1)
    # Use the trained regressor to make a prediction for the current row
    y_pred = regressor.predict(row)
    print(y_pred)
    # Append the predicted value to the list
    y_pred_list.append(y_pred[0])  # The predicted value is a 1-dimensional array with a single element, so we extract the element using indexing

y_pred_array = np.array(y_pred_list)
```

```python
x_test.shape
```

```python
y_test=np.array(y_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

mae = mean_absolute_error(y_test, y_pred_array)

# Calculate the accuracy in terms of percentage
accuracy = (1 - mae / y_test.mean()) * 100

print("Accuracy: {:.2f}%".format(accuracy))
```
